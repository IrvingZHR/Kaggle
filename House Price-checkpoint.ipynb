```python
x = df.loc[:,~df.columns.isin(['Id','SalePrice'])]
y = df.loc[:,['SalePrice']]

x_test = test_df.loc[:,~test_df.columns.isin(['Id','SalePrice'])]
```

```python
ohe = OneHotEncoder(sparse=False)
x_cat_encoded_total = ohe.fit_transform(x_cat_toal)
x_cat_df_total = pd.DataFrame(x_cat_encoded_total, columns = ohe.get_feature_names_out(x_cat_toal.columns))
```

```python
x_cat_df = x_cat_df_total.iloc[:1460,]

x_cat_df_test = x_cat_df_total.iloc[1460:,]
```

```python
x = x_num.join(x_cat_df)

x_test = x_num_test.join(x_cat_df_test)
```

```python
imputer = SimpleImputer(strategy = 'constant', fill_value = 0)

x_impute = imputer.fit_transform(x)


x_impute_test = imputer.fit_transform(x_test)
```

```python
rf = RandomForestRegressor(n_estimators = 300,random_state = 42) 
ad = AdaBoostRegressor(n_estimators = 300,random_state = 42)
xgbr = xgb()
```

```python
kf = KFold(n_splits = 10,shuffle = True, random_state = 42)
rf_score = cross_val_score(rf, x_imputer, y.values.ravel(), cv = kf)
mean_rf_score = np.mean(rf_score)
print(mean_rf_score)
```

```python
kf = KFold(n_splits = 10,shuffle = True, random_state = 42)
ad_score = cross_val_score(ad, x_imputer, y.values.ravel(), cv = kf)
mean_ad_score = np.mean(ad_score)
print(mean_ad_score)
```

```python
kf = KFold(n_splits = 10,shuffle = True, random_state = 42)
xgbr_score = cross_val_score(xgbr, x_imputer, y.values.ravel(), cv = kf)
mean_xgbr_score = np.mean(xgbr_score)
print(mean_xgbr_score)
```

```python
xgbr_fit = xgbr.fit(x_impute,y.values.ravel())
```

```python
y_pred  = np.exp(xgbr_fit.predict(x_impute_test))
```

```python
xgbr_pred_df = pd.DataFrame({'Id': test_df['Id'],'SalePrice': y_pred})
```
